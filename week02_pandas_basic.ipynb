```python
# 1-Dimension

a = np.array([0, 1, 2, 3])

print(a.ndim, a.shape, len(a))
# ndim vs dim(R)
# 간단하지만 헷갈렸는데 ndim은 차원의 수 array, matrix, tensor (1, 2, 3) 표시
# dim은 R에서 사용하는 매서드이고 python에서는 shape과 같은 역할로 배열의 모양 표시 (n by n)
```

```python
# ndim과 shape

# 2행 3열짜리 2차원 배열 생성
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(f"차원의 수 (ndim): {arr.ndim}")  # 결과: 2
print(f"배열의 모양 (shape): {arr.shape}") # 결과: (2, 3)
```

```python
# array 만들기 (arange, linspace, ones, zeros, eye(I), diag)

arr1 = np.arange(10)

arr2 = np.arange(1, 9, 2)

arr3 = np.linspace(0, 1, 6)

arr4 = np.linspace(0, 1, 5, endpoint = False)

arr5 = np.ones((3, 3)) # tuple 대신 list도 들어감 하지만 tuple이 관례이고 shape을 사용했을 때 tuple로 나오는 것처럼 불변적인 tuple이 좋음

arr6 = np.zeros((2,2))

arr7 = np.eye(3)

arr8 = np.diag(np.array([1, 2, 3, 4]))

print(arr1, "\n", arr2, "\n", arr3, "\n", arr4, "\n", arr5, "\n", arr6, "\n", arr7, "\n", arr8)
```

```python
import numpy as np

# 1. 각 행의 시작 숫자(0, 10, 20, 30, 40)를 만듭니다.
# .reshape(5, 1)을 통해 '열 벡터(세로 한 줄)'로 세워줍니다.
row_starts = np.arange(0, 50, 10).reshape(5, 1)

# 2. 각 행에 더해줄 0부터 5까지의 숫자를 만듭니다. (가로 한 줄)
row_values = np.arange(6)

# 3. 넘파이의 '브로드캐스팅' 기능이 세로 줄과 가로 줄을 합쳐 5x6 행렬을 만듭니다.
result = row_starts + row_values

print(result)
```

```python
# rng와 같이 default_rng 사용을 추천
# 이전의 방식은 전체 랜덤 시드를 고정시키는 단점과 미세한 통계적 차이 존재

rng = np.random.default_rng(42)
a = rng.random(4)       # uniform in [0, 1]

b = rng.standard_normal(4)

print(a, b)
```
