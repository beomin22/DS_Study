```python
#기초 확인하기 BMI 계산기 제작

# 환자 데이터 리스트 (이름, 나이, 키(m), 몸무게(kg))
patients = [
    {"name" : "Lee", "age" : 24, "height" : 175, "weight" : 60},
    {"name" : "Nam", "age" : 25, "height" : 173, "weight" : 75},
    {"name" : "Seo", "age" : 50, "height" : 160, "weight" : 90}
]

overweight = []

for patient in patients:
    bmi = patient["weight"] / ((patient["height"] / 100) ** 2)
    patient["bmi"] = round(bmi, 1)
    if patient["bmi"] >= 25:
        overweight.append(patient["name"])

print(f"비만인 사람은 {overweight} 입니다.")
```

```python
# 기초 확인하기 BMI 계산기 제작 함수 버전

# 환자 데이터 리스트 (이름, 나이, 키(cm to m), 몸무게(kg))
patients = [
    {"name" : "Lee", "age" : 24, "height" : 175, "weight" : 60},
    {"name" : "Nam", "age" : 25, "height" : 173, "weight" : 75},
    {"name" : "Seo", "age" : 50, "height" : 160, "weight" : 90}
]

# bmi 함수
def check_bmi(hei, wei):
    bmi = wei / ((hei/100) ** 2)
    return round(bmi, 1)

overweight = []

for patient in patients:
    patient["bmi"] = check_bmi(patient["height"], patient["weight"])
    if patient["bmi"] >= 25:
        overweight.append(patient["name"])

print(f"비만인 사람의 이름은 {overweight}입니다.")
```
